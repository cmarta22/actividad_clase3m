crear un diccionario con informacion de productos (nombre del producto,precio,stok)
utilice el metodo .keys() par obtener una lista de llaves (nombre del producto) en el diccionario
utilice el metodo .values() para obtener una lista de valores (precio,cantidad)en el diccionario
utilice el metodo .items() para obtenr una lista de tuplas con pares clave valor del diccionario
muestre como utilizar el metodo .get() para obtener el precio del producto especifico del diccionario

```python
productos={"lavandina":{"precio":200, "stock":100},
           "jabon":{"precio":300, "stock":50},
           "shampoo":{"precio":400, "stock":20}}
```

```python
print(productos)
```

```python
nombre_producto=list(productos.keys())
```

```python
print(nombre_producto)
```

```python
precio_cantidad=list(productos.values())
```

```python
print(precio_cantidad)
```

```python
clave_valor=list(productos.items())
```

```python
print(clave_valor)
```
